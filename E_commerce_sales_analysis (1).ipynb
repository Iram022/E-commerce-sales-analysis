```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("amazon.csv")
df.head()
```

```python
df.info()
df.describe()
df.isnull().sum()
df.nunique()
```

### Data Cleaning

```python
df.drop_duplicates(inplace=True)
```

```python
df['discounted_price'] = df['discounted_price'].str.replace('₹','').str.replace(',','').astype(float)
df['actual_price'] = df['actual_price'].str.replace('₹','').str.replace(',','').astype(float)
df['discount_percentage'] = df['discount_percentage'].str.replace('%','').astype(float)
```

###### Clean ratings

```python
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df['rating_count'] = df['rating_count'].str.replace(',','').astype(float)
```

```python
df.dropna(subset=['rating'], inplace=True)
```

### EDA

##### Category Distribution

```python
df['category'].value_counts()
```

##### Average discounted price by category

```python
df.groupby('category')['discounted_price'].mean().sort_values()
```

###### Average rating by category

```python
df.groupby('category')['rating'].mean().sort_values(ascending=False)
```

######

```python
df['main_category'] = df['category'].str.split('|').str[0]
```

###### Relationship between discount and rating

```python
sns.scatterplot(data=df, x='discount_percentage', y='rating')
```

###### Top 10 most expensive products

```python
df.sort_values('actual_price', ascending=False).head(10)
```

### Visualization

Category countplot

```python
plt.figure(figsize=(12,6))
sns.countplot(y=df['main_category'], order=df['main_category'].value_counts().index)
plt.title("Product Count by Main Category")
plt.show()
```

Price distribution

```python
sns.histplot(df['discounted_price'], bins=50)
plt.title("Price Distribution")
```

Rating distribution

```python
sns.histplot(df['rating'], bins=20)
plt.title("Rating Distribution")
```

Discount vs rating

```python
sns.scatterplot(x=df['discount_percentage'], y=df['rating'])
```

Correlation Heatmap

```python
plt.figure(figsize=(8,6))
sns.heatmap(df[['discounted_price','actual_price','rating','rating_count','discount_percentage']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Between Numeric Features")
plt.show()
```
